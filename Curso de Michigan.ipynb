**Slicing**

Vamos ao Slicing, que é uma maneira de "cortar" um Array menor de outro maior. Para uma dimensão, esse processo é semelhante ao aplicado nas listas. Usaremos os dois pontos (:) para isso. Por exemplo, se colocarmos ":3" nos indicadores de índice "[]", nós obtemos os elementos do índice zero até o 3. Lembrando que o índice 3 é excluído. Criaremos para visualizarmos um array, de 0 a 5. 6 números. Depois, vamos visualizar: print( a [ :3 ] ) Tendo assim os três primeiros elementos. Se colocarmos [2:4], veremos do segundo elemento ao quarto. Mas, novamente, excluindo o quarto. Então: print( a [ 2 : 4 ] ) Ao invés de ver a distância, vamos ver os elementos entre esses índices. Para matrizes multidimensionais, funciona da mesma forma. Vamos ver um exemplo: a = np.array... Vamos criar uma nova lista. E aqui vamos colocar três listas. Vamos ver a matriz que possui três linhas e quatro colunas. Se colocarmos um argumento em um Array, como a[:2], teremos todos os elementos desde o primeiro (0.º), até os da segunda linha (1.º em ordem). Se adicionarmos outro argumento ao Array, por exemplo, a[:2, 1:3], veremos elementos das duas primeiras linhas, porém restritos aos elementos que são da 2ª e 3ª colunas. Vamos tentar fazer isso: a[:2, 1:3] Então, nós queremos tudo das duas linhas e queremos 1 : 3 (da coluna). Resumindo, em matrizes multidimensionais, o primeiro argumento é para selecionar linhas e o segundo selecionar colunas. É importante perceber que este "corte" em um Array é apenas uma mudança na exibição dos mesmos dados. Isso é realmente importante e é chamado "passagem por referência". Portanto, modificar este sub Array pode modificar o Array original, também. Aqui eu vou mudar o elemento da posição [0,0] que é 2, para 50. Assim veremos que o valor no Array original também mudou para 50. Eu criarei um sub_array e usarei o mesmo processo de antes. sub_array = a[ :2, 1:3 ] Pegamos duas linhas e, depois, "1:3" para selecionarmos as colunas. E mostrarei o valor de sub_array[0, 0]. Então, eu vou estabelecer: sub_array[0, 0] = 50 Eu vou mudar para 50. Então nós vamos mostrar o que esse sub array nos dá. Isso deve ser 50. Mas, então, nós vamos pedir para imprimir o nosso array original. Isso é “a” também. Lembre-se de que aqui, é a[0, 1], porque nós mudamos quais colunas seriam selecionadas para o nosso sub array. Então, quando nós pegamos o nosso sub array e fizemos [1 : 3], nós nos livramos da 0ª coluna, em “a”. Então, a 0ª coluna em sub array é a primeira, ou seja, a coluna 1 em “a”.


```python
a=np.array([0,1,2,3,4,5])
print(a[:3])
```

Regexes não são comuns, mas eles são uma técnica fundamental para a organização ou "limpeza" dos dados em muitas aplicações em Data Science.

```python
import re
```

```python
#Existem várias funções de processamento e leitura importantes que você pode usar. A primeira, match(), checa por uma correspondência entre strings,
#começando desde o início dela, para assim retornar um dado Booleano. Da mesma forma, search(), verifica se há uma correspondência, retornando um Booleano,
#mas podendo pesquisar em qualquer posição.
text = "This is a good day"

if re.search("good",text):
  print("Wonderful!")
else:
  print("Alas :(")
```

```python
#Mas, além de verificar por condicionais, podemos dividir a string. O trabalho que o Regex faz aqui é chamado "Tokenizing",
#onde a string é separada em substrings baseadas em padrões. Tokenizing é uma atividade principal em processamento de linguagem natural,
#que apesar de não falarmos muito sobre isso aqui, você estudará futuramente. Temos também as funções findall() e split(),
#que vão analisar a string para nós e retornar suas partes.
text ="Amy works diligently. Amy gets good grades. Our student Amy is succesful."

re.split("Amy",text)
```

```python
#Para contar quantas vezes Amy aparece no basta usar findall()
re.findall("Amy", text)
#Ok, então nós vimos que Search() procura por algum padrão e retorna um booleano, que Split() usará um padrão para criar uma lista de substrings e que
#Findall() vai procurar um padrão e mostrar todas suas ocorrências. Bem, já sabemos como a API RegEx do Python funciona.
```

```python
#Vamos agora falar um pouco sobre padrões complexos. A especificação padrão do Regex define uma linguagem de marcação para descrever padrões no texto.
#Vamos começar com os "Anchors". Anchors especificam o início e/ou fim da string que estamos tentando corresponder.
#O circunflexo ^ significa o começo e o sinal do cifrão $ significa o fim. Se você colocar o circunflexo antes de uma string, significa que os textos que o processador
#Regex procura deve começar com a string que você especificar. Quando é no fim, você tem que colocar o sinal "$" para identificar o final de uma string alvo.
#Isso significa que os textos que o Regex recupera devem terminar com a string que você especificou.
text ="Amy works diligently. Amy gets good grades. Our student Amy is succesful."
re.search("^Amy",text)
```

**Quantificadores**

```python
#Quantificadores são o número de vezes que você quer buscar um padrão correspondido antes de ser contada a correspondência de fato.
#O Quantificador mais básico é expresso por e{m,n}, com m e n entre chaves, onde "e" é a expressão ou caractere que queremos combinar, "m"
#é o número mínimo de vezes que você quer que ele seja combinado, e "n" é o número máximo de vezes que o item pode ser combinado.
#Vamos usar aquelas notas em 'grades' como exemplo. Quantas vezes este estudante esteve em uma sequência só de notas A? Para vermos isso, faremos o re.findall
#onde o caractere que estamos interessados é A, de modo que ele toma o lugar de e, abriremos as chaves, queremos ter pelo menos dois dele,
#de modo que é o nosso m terá o valor 2, e aqui usaremos apenas algum valor grande, tal como o 10, fechando chaves e usando tudo isso para buscar em "grades".
re.findall("A{2,10}",grades)
#Aqui estamos usando o 2 como nosso mínimo e o 10 como nosso máximo. Vemos então que houve duas sequências, um onde o aluno teve quatro A's e um onde ele só tinha dois A's.
#Podemos tentar fazer isso usando valores únicos e apenas repetindo o padrão. Faremos isso com "re.findall" procurando várias letras A maiúsculas e queremos um mínimo/máximo
#de apenas uma letra A inicial, se repetindo para o segundo caractere e sempre procurando em "grades".
#Como você pode ver, isso é diferente do primeiro exemplo. O primeiro padrão está procurando por uma junção de dois A's, com até dez A's seguidos,
#então ele vê quatro A's como uma única sequência. Já o segundo está procurando por apenas dois A's seguidos, então ele vê os quatro A's como duas sequências de dois A's.
#Dizemos que o processador Regex começa no início da string e utiliza variáveis para corresponder aos padrões que ele faz.
#É importante observar que a sintaxe do Quantificador Regex não permite que você se desvie do padrão {m,n}.
```

```python
#Como podemos trazer vários DataFrame objetos juntos, seja por fusão horizontalmente ou por concatenação verticalmente, intersecção em conjuntos é conhecido como junção em banco de dados.
#Exemplo de como fazer merge
import pandas as pd
#primeiro criar os dois dataframes
staff_df = pd.DataFrame([{'Name': 'Kelly',  'Role': 'Director of HR' },
                         {'Name': 'Sally',  'Role': 'Course Liason'},
                         {'Name': 'James',  'Role':  'Grader'}])
#E vamos indexar os staffs por nome
staff_df = staff_df.set_index('Name')
#Agora vamos criar um dataframe de estudantes
student_df = pd.DataFrame([{'Name': 'James',  'School': 'Business'},
                           {'Name': 'Mike', 'School': 'Law'},
                           {'Name': 'Sally',  'School': 'Engineering'}])
#E agora vamos indexar por nome também
student_df = student_df.set_index('Name')
#agora vamos imprimir o que sai do dataframe
print(staff_df.head())
print(student_df.head())
```

```python
#se quisermos a união desses dfs, chamariamos de mesclagem no DataFrame à esquerda e o DataFrame ligado a combinação certa e reveladora de que a queremos para usar uma junção externa.
#Gostariamos de usar os indices esquerdos e direito como colunas de união.
pd.merge(staff_df, student_df, how='outer', left_index=True, right_index=True)
```

```python
#Vemos no DataFrame resultante que o mike não tem uma função e John não tem um curso, essas células estão listadas como valores ausentes
#Se quisermos uma mesclagem que apenas contenham aquelas pessoas que são um aluno e um membro da equipe podemos definir como atributo a ser inserido intenamente
pd.merge(staff_df, student_df, how='inner', left_index=True,  right_index=True)
#Podemos ver que o resultado do DataFrame só tem James e Sally nele. Agora temos apenas dois usos comuns
#quando juntamos DataFrames, em ambos exempols nós podemos chamar de adição. O primeiro é quando nos queremos todos os staff sem considerar se são estudantes ou não.
#É importante notar nessa função que: o primeiro dataframe pe o esquerdo e o segundo é o direito.
```

**Splitting**

```python
#vamos ver um exemplo
import pandas as pd
import numpy as np
df = pd.read_csv('datasets/census.csv')
#E excluir resumos em nível estadual, que tem o valor menor que 40
df = df[df['SUMLEV']==50]
df.head()
```

```python
#no primeiro exemplo para groupby() queremos usar o dado de censo. Vamos obter uma lista de estados exclusivos, e em seguida
#podemos iterar sobre todos esses estados. Para cada estado, podemos reduzir o quadro de dados e calcular a média.

#Vamos executar essa tarefa três vezes e cronometrá-la. Para isso, vamos usar essa função mágica celular %%timeit
```

```python
%%timeit -n 3
for state in df['STNAME'].unique():
  #Vamos usar o NumPy para calcular esse estado específico.
  avg = np.average(df.where(df['STNAME']==state).dropna()['CENSUS2010POP'])
  print('Countries in state' +state +
        'have an average population of '+ str(avg))
```

```python
#Agora vamos tentar uma abordagem tentando groupby()
```

```python
%%timeit -n 3
#Para esse método,começando dizendo ao Pandas que estamos interessados em agrupar o estado por
#nome. Essa é a nossa divisão.
for group. frame in df.groupby('STNAME'):
  #É notavel que há dois valores que definimos aqui. Na verdade, Groupby retorna uma tupla
  #em que o primeiro valor é o valor da chave pela qual estamos tentando agrupar, neste caso
  #um nome de estado específico. O segundo é o quadro de dados projetados que foi encontrado para esse grupo.
  #Agora incluimos nossa lógica na etapa de aplicação, que é apenas para calcular uma média do
  #Census2010POP
  avg = np.average(frame['CENSU2010POP'])
  print('Countries in state' + group +
        'have an average population of '+ str(avg))
  # Pode fornecer uma função para groupby e usa-lá para segmentar seus dados. Exemplo
  #É importante observar que, para fazer isso, você precisa definir o índice do quadro de dados como a coluna pela qual você deseja agrupar primeiro.
  #Criaremos uma nova função chamada set_batch_number. Se o primeiro número do parâmetro for M maiúsculo, retornaremos um zero. Se for Q maiúsculo, retornaremos um.
  #Caso contrário, devolveremos dois. Em seguida, passaremos essa função para o DataFrame.
  df = df.set_index('STNAME')
  def set_batch_number(item):
    if item[0]<'M':
      return 0

    if item[0]<'Q':
      return 1
    return 2

for group, frame in df.groupby(set_batch_number):
  print('There are +' str(len(frame))+ 'records in group'+ str(group)+ 'for processing')

```

```python
#função que gerencia os agrupamentos
def grouping_fun(item):
    #CVerificaremos se a parte review_scores_value do índice. O item está no formato da política de cancelamento e review_scores_value.
    #Então é uma tupla. Se o item sub-um for igual a 10, retornaremos os itens abaixo de zero e 10,0.
    if item[1] == 10.0:
      return(item[0],"10.0")
    else:
      return (item[0],"not 10.0")

for group, frame in df.groupby(by=grouping_fun):
  print(group)
```

**Aggregation**

```python
#A etapa de aplicação mais simples é a agregação de dados. Isso usa um método chamado agg no objeto groupby.
#Até agora, nós apenas iteramos o objeto groupby, desempacotando-o em um rótulo, o nome do grupo e um DataFrame.
#Mas com agg, podemos passar um dicionário das colunas que estamos interessados em agregar junto com a função que queremos aplicar.
df=df.reset_index()
#exemplo
df.groupby("cancellation_policy").agg({"review_scores_value":np.average})
```

```python
df.groupby("cancellation_policy").agg({"review_scores_value"}:np.nanmean)
#Podemos também estender esse dicionário to aggregate para várias funções ou múltiplas colunas,
df.groupby("cancelation_policy").agg({"review_scores_value":(np.nanmean,np.nanstd),"reviews_per_month":np.nanmean})
```

**Transformation**

```python
#A transformação é diferente das agregações. Onde agg retorna um único valor por coluna, ou seja, uma linha por grupo, transform retorna um objeto do mesmo tamanho do grupo.
#Essencialmente, ele transmite a função que você fornece pelo dataframe do grupo, retornando um novo dataframe. Isso facilita bastante a combinação de dados posteriormente.
#Por exemplo, suponha que quiséssemos incluir os valores médios de classificação em um determinado grupo por política de cancelamento, mas preservar as formas do quadro de dados
#para que pudéssemos gerar uma diferença entre uma observação individual e a soma.
#Primeiro, vamos definir um subconjunto das colunas nas quais estamos interessados.
#Aqui, vou criar uma coluna variável e dizer que estamos interessados na política de cancelamento e no valor das pontuações das avaliações,
#o que só torna tudo um pouco mais limpo enquanto eu o digito. Agora vamos transformá-lo. Vou armazenar isso em seu próprio dataframe.
#Exemplo
cols=['cancellation_policy','review_score_value']
#Agora vamos transformar,Irá armazenar no próprio dataframe
transform_df=df[cols].groupby('cancellation_policy').transform(np.nanmean)
transform_df.head()
```

```python
transform_df.rename({'review_scores_value':'mean_review_scores'},axis='colums', inplace=True)
df=df.merge(transform_df, left_index=True, right_index=True)
df['mean_diff']=np.absolute(df['review_scores_value']-df['mean_review_scores'])
df['mean_diff'].head()
```

**Filtering**

```python
# Portanto, o objeto Group-by também é compatível com a filtragem de grupos.
#Geralmente, você deseja agrupar por alguns recursos, fazer algumas transformações nos grupos e, em seguida, eliminar determinados grupos como parte de sua rotina de limpeza.
#A função de filtro assume uma função que se aplica a cada quadro de dados do grupo e retorna um verdadeiro ou falso, dependendo se esse grupo deve ser incluído nos resultados.
#Por exemplo, se quiséssemos, apenas os grupos que têm uma classificação média acima de nove, incluídos em nossos resultados.
df.groupby('cancellation_policy').filter(lambda x: np.nanmean(x['review_scores_value'])>9.2)
#Observe que os resultados ainda estão indexados, mas que qualquer um dos resultados que estavam no grupo com a pontuação média da revisão menor ou igual a 9,2.
#Não somos copiados por causa de como fizemos nossa projeção.
```

**Tipos de Escala e dados**

#Escala de Proporção
Na escala de proporção, as unidades de medida são igualmente espaçadas e operações matemáticas como a subtração, divisão e multiplicação são todas válidas. Bons exemplos de escalas de proporção são altura e peso.
#Escala de Intervalo
Na escala de intervalo, as unidades de medida são igualmente espaçadas, como na escala de proporção. Mas não há ausência clara de valor. Ou seja, não há um verdadeiro zero, e assim operações tais como multiplicação e divisão não são válidas.
Um exemplo da escala de intervalos pode ser a temperatura medida em graus Celsius ou Fahrenheit uma vez que nunca há uma ausência de temperatura, e zero graus é, na verdade, um valor significativo em si. A direção em uma bússola pode ser outro bom exemplo, onde zero graus na bússola não indica uma falta de direção, mas em vez disso, descreve uma direção em si. Para a maior parte do trabalho que você fará com mineração de dados, as diferenças entre intervalos de proporção intervalos de escala, podem não ser claramente visíveis ou importantes para o algoritmo que você aplicará, mas é importante ter uma distinção clara em sua mente ao aplicar testes estatísticos avançados.
#Escala Ordenada
A ordem dos valores é importante, mas a diferença entre os valores não é igualmente espaçada. Exemplo notas em letras como A, A+, quando você compara com um valor percentual, a letra por si só cobre 4% dos graus disponíveis e uma letra com um sinal é usada em apenas 3% das notas disponíveis.
Os dados ordenados são muito comuns em 'aprendizagem de máquina' e podem, algumas vezes, ser um trabalho desafiador.
#Escala Nominal
muitas vezes chamadas dados categóricos. Aqui nomes de times em um esporte podem ser um bom exemplo. Há um limitado número de times, mas mudar a sua ordem ou aplicar funções matemáticas a eles não faz sentido. Valores categóricos são muito comuns e, geralmente, nos referimos a dados categóricos onde há apenas dois valores possíveis, tal como categorias binárias

```python
df = pd.read_csv('datasets/cwurData.csv')
#nesse video foi criado uma função para classificar e separar as melhores universidades do mundo em tier, do 1 ao 100, do 101 ao 200, etc...
def create_category(ranking):
  if (ranking >= 1) & (ranking <= 100):
    return "First Tier Top University"
  elif(ranking >= 101) & (ranking <= 200):
    return "Second Top Tier University"
  elif(ranking >= 201) & (ranking <= 300):
    return "Third Top Tier University"
#Agora podemos aplicar aplicar a uma única coluna para criar uma nova série
df['Rank Level']= df['world_rank'].apply(lambda x: create_category(x))
df.head()
```

**Testes estatísticos básicos**

```python
#Utilizaremos o SciPy para executar o teste t de student, usamos muita estatística de diversas
#formas em ciência de dados, nessa aula focaremos em teste de hipóteses, que é uma atividade
#central na análise de dados, por trás de diversos experimentos.
#O objetivo do teste de hipóteses é determinar, se por exemplo, as duas diferentes condições
#que temos em nosso exérimento resultaram em impactos distintos.
import numpy as np
import pandas as pd
#Agora vamos traze novas bibliotecas do SciPy.
from scipy import stats
```

```python
#Scipy é uma coleção interessante de bibliotecas para ciência de dados, Matplotlib é uma biblioteca
#de gŕaficos.
#Quando fazemos um teste de hipóteses temos duas afirmações de interesse. A primeire á nossa
#explicação real, que chamamos de hipótese alternativa, e a segunda é a explicação que temos
#que não é suficiente, e nós a chamamos de hióteses nula.
```

```python
#Exemplo
df=pd.read_csv('datasets/grades.csv')
df.head()
#veremos 6 tarefas diferentes(assingnments 1,2,..6), com a nota ("grade") e a data de submissão de cada uma.
print("There are {} rows and {} columns".format(df.shape[0], df.shape[1]))
```

```python
early_finishers=pd.df[pd.to_datatime(df['assignment1_submission'])<'2016']
early_finishers.head()
```

```python
late_finishers=df[~df.index.isin(early_finishers.index)]
late_finishers.head()
```

```python
#Como visto o objeto DataFrame do pandas tem uma variedade de funções estatísticas associacdas
#se chamar a função de média "mean" diretamente no DataFrame, verá que todas as médias para as
#tarefas foram calculadas.
print(early_finishers['assignment1_grade'].mean())
print(late_finishers['assignment1_grade'].mean())
```

```python
#Teste T de Student Ele nos permite formular a hipótese alternativa
#de que as médias são diferentes, e também a hipótese nula
#de que as médias são iguais, e então testar a hipótese nula.
#Quando fazemos um teste de hipóteses,temos que escolher um nível
#de significância como um limite de quão possível estamos dispostos
#a aceitar. Este nível de significância é tipicamente chamado de alfa.
#EXEMPLO: um limite de 0.05 para nosso alfa, que é cinco porcento.
#Usaremos a função ttest_ind(), que realiza um teste t independente.
```

```python
from scipy.stats import ttest_ind
ttest_ind(early_finishers['assignment1_grade'], late_finishers['assignment1_grade'])
```

```python
#Vemos que a possibilidade é 0,18. Isso significa que não podemos rejeitar
#a hipótese nula. A hipótese nula foi que as duas populações eram iguais.
#Não temos certeza o suficiente em nossa evidência porque a probabilidade é
#maior que alfa para chegar a uma conclusão contrária. Isso não significa que
#provamos que as populações são iguais.
```

```python
print(ttest_ind(early_finishers['assignment2_grade'], late_finishers['assignment2_grade']))
print(ttest_ind(early_finishers['assignment3_grade'], late_finishers['assignment3_grade']))
print(ttest_ind(early_finishers['assignment4_grade'], late_finishers['assignment4_grade']))
print(ttest_ind(early_finishers['assignment5_grade'], late_finishers['assignment5_grade']))
print(ttest_ind(early_finishers['assignment6_grade'], late_finishers['assignment6_grade']))
```
