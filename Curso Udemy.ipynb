**Aula 3 Estrutura de Repetição**

```python
count = 1
while count <=5:
    print(count)
    count ++ 1
```

```python
for n in range(0, 10):
    print(n+1)
for n in range(1, 11):
    print(n)
```

```python
for n in range(10 , 0 -1):
    print(n)
    
```

```python
#exemplo com break
for n in range(0, 10):
    if n == 4:
        break
    print(n)
```

```python
#exemplo com continue
for n in range(0, 10):
    if n == 4:
        continue
    print(n)
```
